```python
import numpy as np

img=cv2.imread("q2.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,2,300,
                            param1=50,param2=30,minRadius=0,maxRadius=0)


circles = np.uint8(np.around(circles))

for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(gray,(i[0],i[1]),i[2],(0,255,0),2)
   
    plt.imshow(gray)
```

```python
img=cv2.imread("q3.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
final_img = img
img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

img_blur = cv2.GaussianBlur(img , (5,5), 0)

img_edges = cv2.Canny(img_blur, 40, 200, 3)

img_blend = cv2.addWeighted(img_edges,0.5,img_blur,0.1,0)

threshold_value , img_gray  = cv2.threshold (img_blend ,127 , 255 , cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(img_gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for i in range (len(contours)):
    cv2.drawContours (final_img ,contours,i,(0,0,255),1)
final_img = cv2.cvtColor (final_img  , cv2.COLOR_BGR2RGB)        
plt.imshow (final_img)
```

```python
print ("Changing color points Will be: " + str(len (contours) ))
```
